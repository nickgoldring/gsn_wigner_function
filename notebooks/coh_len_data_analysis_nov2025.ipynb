```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.special as spec
```

```python
# drift length values [m]
L_vals = np.array([0.25, 0.5,	1,	2,	3,	7,	9,	15])

# coherence length values [um]
analytic_vals = np.array([7.1, 8.4,	12.4,	22,	32,	74,	95,	158])
gsn_fit_vals = np.array([9.38, 7.6,	14.8,	29,	44,	102,	132,	220])
fwhm_vals = np.array([5.27, 9.8,	19.5,	39,	58,	135,	174,	290])
gsn_fit_sqrt2_vals = gsn_fit_vals / np.sqrt(2)

fwhm_fact = 1.84       # correction factor for FWHM/analytic at far field

fwhm_correct_vals = fwhm_vals / fwhm_fact
```

```python
plt.plot(L_vals, analytic_vals, 'b--', label='analytic')
plt.plot(L_vals, gsn_fit_vals, 'r--', label='Gaussian fit')
plt.plot(L_vals, gsn_fit_sqrt2_vals, 'm:', label='Gaussian fit / sqrt(2)')
plt.plot(L_vals, fwhm_vals, 'g-', label='FWHM')
plt.plot(L_vals, fwhm_correct_vals, 'k.', label='FWHM / 1.84')

# plt.xlim(-theta[-1], theta[-1])
# plt.ylim(0.0, 1.1 *P_theta[0]) 
plt.legend()
plt.show()
```

```python
plt.plot(L_vals, analytic_vals, 'b.-', label='analytic')
plt.plot(L_vals, gsn_fit_sqrt2_vals, 'm.-', label='Gaussian fit / sqrt(2)')
# plt.xlim(-theta[-1], theta[-1])
# plt.ylim(0.0, 1.1 *P_theta[0])
plt.legend()
plt.show()
```
